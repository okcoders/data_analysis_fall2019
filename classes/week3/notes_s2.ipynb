```python
# Subset columns, specifying order
cars[["mpg", "wt", "gear"]].head()
```

```python
cars.mpg > 20
```

```python
cars[cars.mpg > 20]
```

```python
# Note: it is very important that you have parentheses around each logical test
cars[(cars.mpg > 20) & (cars.gear == 4)]
```

```python
# Note: without the parens it will fail
cars[cars.mpg > 20 & cars.gear == 4]
```
