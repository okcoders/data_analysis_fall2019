### Merging

```python
import pandas as pd
import sqlite3
```

```python
emp_db = sqlite3.connect("../../data/company_employees_plus.db")

employees = pd.read_sql("SELECT * FROM employees", con = emp_db)
countries = pd.read_sql("SELECT * FROM country_master", con = emp_db)
continents = pd.read_sql("SELECT * FROM continent_master", con = emp_db)

print(len(employees), len(countries), len(continents))
```

```python
employees.merge(right = countries, on = "country", how = "inner").head()
```

```python
(countries
    .merge(
        right = continents.assign(continent = continents.name),
        on = "continent",
        how = "inner"
    )
)
```

```python
(countries
    .merge(
        right = continents.rename(columns = {"name": "continent"}),
        on = "continent",
        how = "inner"
    )
)
```

```python
(countries
    .rename(columns = {"population": "pop_country"})
    .merge(
        right = continents.rename(columns = {"name": "continent", "population": "pop_continent"}),
        on = "continent",
        how = "inner"
    )
)
```

```python
(countries
    .rename(columns = {"population": "pop_country"})
    .merge(
        right = continents.rename(columns = {"population": "pop_continent"}),
        left_on = "continent",
        right_on = "name",
        how = "inner"
    )
)
```

```python
(countries
    .rename(columns = {"population": "pop_country"})
    .merge(
        right = continents.rename(columns = {"population": "pop_continent"}),
        left_on = "continent",
        right_on = "name",
        how = "right"
    )
)
```

```python
employees.rename(columns = {"email": "steve", "total_worth": "n_motorcycles"}).head()
```

```python
employees.drop(columns = ["company", "email", "gender"]).head()
```

```python
agg_comp = (employees
    [["company", "total_worth"]]
    .groupby("company").mean()
    .rename(columns = {"total_worth": "avg_worth"})
)


merged = (employees
    .drop(columns = ["is_management", "email", "country", "gender"])
    .merge(
        right = agg_comp,
        on = "company",
        how = "inner"
    )
)

merged[(merged.total_worth > merged.avg_worth) & (merged.avg_worth > 40)].head(10)
```

```python
agg_comp = (employees
    [["company", "total_worth"]]
    .groupby("company").mean()
    .rename(columns = {"total_worth": "avg_worth"})
)


merged = (employees
    .drop(columns = ["is_management", "email", "country", "gender"])
    .merge(
        right = agg_comp[agg_comp.avg_worth > 40],
        on = "company",
        how = "inner"
    )
)

merged[(merged.total_worth > merged.avg_worth)].head(10)
```
