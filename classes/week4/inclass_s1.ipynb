### Aggregations

```python
cars = ds.data("mtcars")
cars.head()
```

```python
type(cars)
```

```python
type(cars.groupby("cyl"))
```

```python
cars.groupby("cyl").size()
```

```python
cars.groupby("am").size()
```

```python
cars.groupby("cyl").mean()
```

```python
cars.groupby("cyl").mean()[["hp"]]
```

```python
cars.groupby("cyl")[["hp"]].mean()
```

```python
cars.groupby("cyl")[["hp", "mpg"]].mean()
```

```python
output = (cars
    [cars.hp > 75]
    .groupby("cyl")
    [["hp", "mpg"]]
    .mean()
)
output
```
