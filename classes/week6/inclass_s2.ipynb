```python
import pandas as pd
import pydataset as ds
from sklearn.linear_model import LinearRegression
```

```python
boston = ds.data('Boston')
boston["rm_age"] = (boston.rm * boston.age)
len(boston)
```

```python
boston.head()
```

```python
boston
```

```python
model = LinearRegression()
model.fit(X = boston[["chas"]], y = boston.medv)
print(model.score(X = boston[["chas"]], y = boston.medv))
print(model.coef_)
```

```python
model = LinearRegression()
model.fit(X = boston.drop(columns = ['medv']), y = boston.medv)
model.score(X = boston.drop(columns = ['medv']), y = boston.medv)
```

```python
boston.plot.scatter(x = "ptratio", y = "medv", figsize = (16, 8))
```

```python
faithful = ds.data('faithful')
faithful.head()
```

```python
faithful.plot.scatter(x = "waiting", y = "eruptions", figsize = (16, 8))
```

```python
mod_f = LinearRegression()
mod_f.fit(X = faithful[["eruptions"]], y = faithful.waiting)
print(mod_f.score(X = faithful[["eruptions"]], y = faithful.waiting))
print(mod_f.coef_, mod_f.intercept_)
```

```python
mod_f = LinearRegression()
mod_f.fit(X = faithful[["waiting"]], y = faithful.eruptions)
print(mod_f.score(X = faithful[["waiting"]], y = faithful.eruptions))
print(mod_f.coef_, mod_f.intercept_)
5 / mod_f.coef_[0] + mod_f.intercept_
```
